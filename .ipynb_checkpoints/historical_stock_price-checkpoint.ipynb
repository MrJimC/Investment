```python
import requests, json
import numpy as np
import pandas as pd
import yfinance as yf

import matplotlib.pyplot as plt
import numpy as np
```

```python
Tesla’s production depends on battery technology and raw material supply:
	•	Panasonic (6752.T, PCRFY) – Tesla’s long-time battery supplier.
	•	CATL (300750.SZ) – World’s largest EV battery maker (China).
	•	LG Energy Solution (373220.KQ) – South Korean battery giant.
	•	Albemarle (ALB) – Leading lithium supplier for EV batteries.
	•	Piedmont Lithium (PLL) – Lithium mining company with Tesla contracts.
```

```python
# import yfinance as yf

date = '2022-01-01'
stock_no = ['TSLA', 'PCRFY', 'ALB', 'PLL']
stock_history_price = pd.DataFrame()

for stock in stock_no:
    get_stock = yf.Ticker(stock)
    stock_data = get_stock.history(start=date)
    stock_data = stock_data.reset_index()
    stock_data['ticker'] = stock
    stock_history_price = pd.concat([stock_history_price, stock_data], ignore_index = True)

stock_history_price
```

```python
# Create the plot
plt.figure(figsize=(10, 5))

## ['TSLA', 'BYDDF', 'RIVN', 'LCID', 'NIO', 'XPEV', 'GM', 'F']
# Plot different stocks
plt.plot(stock_history_price[stock_history_price.ticker == 'TSLA']['Date'], stock_history_price[stock_history_price.ticker == 'TSLA']['Close'], label='TSLA', color='blue', linestyle='-')
plt.plot(stock_history_price[stock_history_price.ticker == 'PCRFY']['Date'], stock_history_price[stock_history_price.ticker == 'PCRFY']['Close'], label='BYDDF', color='red', linestyle='-')
plt.plot(stock_history_price[stock_history_price.ticker == 'ALB']['Date'], stock_history_price[stock_history_price.ticker == 'ALB']['Close'], label='RIVN', color='green', linestyle='-')
plt.plot(stock_history_price[stock_history_price.ticker == 'PLL']['Date'], stock_history_price[stock_history_price.ticker == 'PLL']['Close'], label='LCID', color='orange', linestyle='-')
# plt.plot(stock_history_price[stock_history_price.ticker == 'NIO']['Date'], stock_history_price[stock_history_price.ticker == 'NIO']['Close'], label='NIO', color='purple', linestyle='-')
# plt.plot(stock_history_price[stock_history_price.ticker == 'XPEV']['Date'], stock_history_price[stock_history_price.ticker == 'XPEV']['Close'], label='XPEV', color='brown', linestyle='-')
# plt.plot(stock_history_price[stock_history_price.ticker == 'GM']['Date'], stock_history_price[stock_history_price.ticker == 'GM']['Close'], label='GM', color='Magenta', linestyle='-')
# plt.plot(stock_history_price[stock_history_price.ticker == 'F']['Date'], stock_history_price[stock_history_price.ticker == 'F']['Close'], label='F', color='black', linestyle='-')


# Customize the chart
plt.xlabel("Date")
plt.ylabel("Stock Price")
plt.title("Stock Price Trend Over Time")
plt.legend()
plt.xticks(rotation=45)  # Rotate X-axis labels for better readability
plt.grid(True)

# Show the plot
plt.show()
```
