```python
import requests, json
import numpy as np
import pandas as pd
import yfinance as yf
```

```python
# import yfinance as yf

date = '2023-01-01'
stock_no = ['0050.TW', '0056.TW']
stock_history_price = pd.DataFrame()

for stock in stock_no:
    stock = yf.Ticker(stock)
    stock_data = stock.history(start=date)
    stock_data['ticker'] = stock
    stock_history_price = pd.concat([stock_history_price, stock_data])

stock_history_price = stock_history_price.reset_index()
stock_history_price
```
